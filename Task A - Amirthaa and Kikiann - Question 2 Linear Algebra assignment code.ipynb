```python
@:Author: Kikiann

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib inline #to plot the charts right into jupyter notebook

#import data set
data = pd.read_csv('C:/Users/kikia/PycharmProjects/pythonProject/2020-21-Intro-to-SSE-dataset-Assignment-2-LA (3) (5).csv')
print(data.shape) #shows the shape of the dataset
print(data.head()) #shows titles of the colums of the dataset with some data

#computing X and Y, x = independent variable, y = dependent variable
x = data['Temperature (degree Celsius)'].values
y = data['Output Raspberry Pie'].values

plt.scatter(x,y)

#Linear Regression (Line fitting)
#get the coefficients
coefficient = np.polyfit(x, y, 1) #number 1 defines the degree of polynomial (first order)
coefficient2 = np.polyfit(x, y, 2) #number 2 defines the degree of the polynomial (second order)
coefficient3 = np.polyfit(x, y, 3) #number 3 defines the degree of the polynomial (third order)

predict = np.poly1d(coefficient)
predict2 = np.poly1d(coefficient2)
predict3 = np.poly1d(coefficient3)

print(predict)
print(predict2)
print(predict3)

#get R2-score from linear model
model = np.poly1d(np.polyfit(x, y, 1))
model2 = np.poly1d(np.polyfit(x, y, 2))
model3 = np.poly1d(np.polyfit(x, y, 3))

from sklearn.metrics import r2_score #python package used to calculate the r2_score of each model

print(r2_score(y, model(x))) #shows the r2_score for the first model
print(r2_score(y, model2(x))) #shows the r2_score for the second model
print(r2_score(y, model3(x))) #shows the r2_score for the third model

#Plot the linear regression model

myline = np.linspace(1, 44, 100) #specifies how the line will display. It will start at 1 and ends at 44
plt.scatter(x, y)   
plt.plot(myline, model(myline))
plt.show()

#plot the polynomial of second order regression model
#draw original scatter plot
plt.scatter(x, y)

#draw line of polynomial regression:
plt.plot(myline, model2(myline))
plt.show()

#plot the polynomial of third order regression model
#draw original scatter plot
plt.scatter(x, y)

#draw line of polynomial regression:
plt.plot(myline, model3(myline))
plt.show()

```
